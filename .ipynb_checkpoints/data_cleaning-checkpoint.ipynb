```python
def transformType(dataframe, fields):
#     type_ = new_df.application_type.unique()
    final = []
    for f in fields:
        type_ = dataframe[f].unique()
        result = dict()
        for i in range(len(type_)):
            result[type_[i]] = result.get(type_[i],i)
        final.append(result)
    return final
fields = ['application_type','class_of_admission','have_agent',"country_of_citizenship",
                   "employer_name","employer_num_employees",'foreign_worker_info_city','foreign_worker_info_education',
                   'foreign_worker_info_major','fw_info_req_experience','fw_info_training_comp',"fw_info_yr_rel_edu_completed",
                   'fw_info_birth_country','fw_ownership_interest','ji_fw_live_on_premises','job_info_foreign_lang_req',
                   'job_info_job_title', 'job_info_major','job_info_training', 'job_info_work_city','job_info_experience',
                   'naics_us_title','preparer_info_title','pw_amount_9089','pw_determ_date', 'pw_expire_date', 
                    'pw_job_title_908','pw_job_title_9089','pw_soc_title','pw_source_name_9089','pw_unit_of_pay_9089',
                   "schd_a_sheepherder",'employer_decl_info_title','case_received_date']
fields = df.columns
fields
```

```python
def getProb(A,B):
    # A is certified and B is other atrribute
    total = A.count()
    probs = []
    for i in B:
        uniq = i.unique()
        prob = []
        for t in uniq:  
            p_cert_and_type = sum(A[i==uniq[t]])/total
            p_t = sum(i==uniq[t])/i.shape[0]
            p_cert_given_type = p_cert_and_type/p_t
            prob.append(p_cert_given_type)
        probs.append(prob)
    return probs
    
# sigma_x - mean_x , sigma_y - mean y
# prob = getProb(df.Certified, newData)
# prob
df.columns
```
